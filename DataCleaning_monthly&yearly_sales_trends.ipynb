LOADING AND READING DATASET

```python
# Import Pandas as pd
import pandas as pd

# Load dataset
file_path = r'C:\Users\SusannaBeatson\Desktop\My Projects\Python - Superstore dataset\Global_Superstore.xlsx'
orders = pd.read_excel(file_path, sheet_name='Orders')
returns = pd.read_excel(file_path, sheet_name= 'Returns')
people = pd.read_excel(file_path, sheet_name= 'People')

orders.head()
#returns.head()
#people.head()
```

CHECKING DATASET STATS AND DESCRIPTION

```python
# Check stats
orders.info()
orders.describe()
```

CHECKING FOR MISSING VALUES

```python
# Check for missing values
orders.isnull().sum()
```

EXTRACTING MONTH AND YEAR FROM DATE

```python
# Create new columns for Year and Month
orders['Year'] = orders['Order Date'].dt.year
orders['Month'] = orders['Order Date'].dt.to_period('M')  # YYYY-MM format

orders.head()
```

GROUPING AND AGGREGATION

```python
# Group data by month column and sum sales for each month
monthly_sales = orders.groupby('Month')['Sales'].sum().reset_index()
monthly_sales.head
```

PLOTTING LINE GRAPH VISUALIZING MONTHLY SALES OVER TIME

```python
# Plot monthly sales trend
import matplotlib.pyplot as plt 

# Convert month to timestamp
monthly_sales['Month'] = monthly_sales['Month'].dt.to_timestamp()

# Calculate average monthly sales
avg_sales = monthly_sales['Sales'].mean()

# Find max and min sales rows
max_row = monthly_sales.loc[monthly_sales['Sales'].idxmax()]
min_row = monthly_sales.loc[monthly_sales['Sales'].idxmin()]

# Plot
plt.figure(figsize=(12, 6))  
plt.plot(monthly_sales['Month'], monthly_sales['Sales'], marker='o', linestyle='-') 
plt.axhline(y=avg_sales, color='blue', linestyle='--', label=f'Average: {avg_sales:.2f}')

plt.title('Monthly Sales Over Time')  
plt.xlabel('Month')  
plt.ylabel('Total Sales')  
plt.grid(True) 

# Annotate highest month
plt.annotate(f'Peak: {max_row["Sales"]:.2f}',
             xy=(max_row['Month'], max_row['Sales']),
             xytext=(max_row['Month'], max_row['Sales'] + 1000),
             arrowprops=dict(facecolor='green', shrink=0.05),
             ha='center', fontsize=10, color='green')

# Annotate lowest month
plt.annotate(f'Low: {min_row["Sales"]:.2f}',
             xy=(min_row['Month'], min_row['Sales']),
             xytext=(min_row['Month'], min_row['Sales'] - 2000),
             arrowprops=dict(facecolor='red', shrink=0.05),
             ha='center', fontsize=10, color='red')


plt.legend()
plt.tight_layout()  
plt.show()  
```

LINE GRAPH VISUALIZING YEARLY SALES OVER TIME

```python
# Step 1: Group by year
yearly_sales = orders.groupby('Year')['Sales'].sum().reset_index()

# Step 2: Calculate average sales
avg_yearly_sales = yearly_sales['Sales'].mean()

# Step 3: Identify peak and low years
max_row = yearly_sales.loc[yearly_sales['Sales'].idxmax()]
min_row = yearly_sales.loc[yearly_sales['Sales'].idxmin()]

# Step 4: Plot with enhancements
plt.figure(figsize=(10, 6))
plt.plot(yearly_sales['Year'], yearly_sales['Sales'], marker='o', linestyle='-')
plt.axhline(y=avg_yearly_sales, color='blue', linestyle='--', label=f'Avg: {avg_yearly_sales:.2f}')

# Annotate peak year
plt.annotate(f'Peak: {max_row["Sales"]:.2f}',
             xy=(max_row['Year'], max_row['Sales']),
             xytext=(max_row['Year'], max_row['Sales'] + 10000),
             arrowprops=dict(facecolor='green', shrink=0.05),
             ha='center', fontsize=10, color='green')

# Annotate lowest year
plt.annotate(f'Low: {min_row["Sales"]:.2f}',
             xy=(min_row['Year'], min_row['Sales']),
             xytext=(min_row['Year'], min_row['Sales'] - 15000),
             arrowprops=dict(facecolor='red', shrink=0.05),
             ha='center', fontsize=10, color='red')

# Labels and grid
plt.title('Yearly Sales Trend')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```
