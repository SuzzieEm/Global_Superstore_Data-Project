PROJECT SUMMARY: SUPERSTORE SALES ANALYSIS


OVERVIEW


This project explores sales and profit trends using the Global Superstore dataset. The analysis was performed using Python, with a focus on extracting actionable insights from customer segments, regions, product categories, and time trends.

The goal was to uncover performance patterns and support strategic decision-making through data visualization and storytelling.


DATA PREPARATION


Cleaned and explored raw order data

Validated data types, handled missing values

Created time-based features (Month, Year) to support trend analysis


KEY INSIGHTS


CUSTOMER SEGMENTS


The Corporate segment outperformed others in total sales and had the highest profit margin, making it the most financially effective customer group.

Consumer and Home Office segments followed, with Home Office maintaining consistent performance despite smaller sales volumes.


SALES TRENDS OVER TIME


Sales fluctuated month-to-month with noticeable peaks, suggesting potential seasonal trends or promotional periods.

A time series analysis highlighted Corporate's sustained lead and visualized trends across all segments.


REGIONAL AND CATEGORY PERFORMANCE


Sales and profits varied by region. Clustered bar charts revealed areas where high sales didn’t always equate to high profit.

All product categories showed healthy profit margins, with no immediate underperformers. Top-selling and most profitable products were identified for each.
